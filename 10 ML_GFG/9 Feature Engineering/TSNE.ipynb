## Loading Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Loading Data

```python
data = pd.read_csv("bank_note_data.csv")
```

```python
data.head()
```

```python
data['Class'].unique()
```

```python
data.info()
```

```python
features = list(data.columns)[:-1]
target = list(data.columns)[-1]
```

## Scaling data

```python
from sklearn.preprocessing import StandardScaler

data_norm = data.copy()
sc = StandardScaler()
data_norm[features] = sc.fit_transform(data[features])
```

```python
data_norm
```

## TSNE

```python
from sklearn.manifold import TSNE

tsne = TSNE(learning_rate=500, n_components = 2)

x_tsne = tsne.fit_transform(data_norm[features])
```

```python
x_tsne
```

```python
y = data[target]
```

```python
y
```

## Visualisation

```python
plt.figure(figsize=(16,11))
# all the x_tsne where class=0
plt.scatter(x_tsne[np.where(y==0), 0], x_tsne[np.where(y==0), 1], color = 'g', marker = "^")
# all the x_tsne where class=1
plt.scatter(x_tsne[np.where(y==1), 0], x_tsne[np.where(y==1), 1], color = 'r', marker = "*")

plt.xlabel("Dimension_1")
plt.ylabel("Dimension_2")
plt.title("TSNE Plot")
plt.show()
```
