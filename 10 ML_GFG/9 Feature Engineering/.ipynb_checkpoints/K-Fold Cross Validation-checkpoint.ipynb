## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.axes as ax

import warnings
warnings.filterwarnings("ignore")
```

## Loading Data

```python
data =  pd.read_csv('boston_house_prices.csv',skiprows=1)
```

```python
data.head()
```

## Splitting the data

```python
x = data.drop(['MEDV'], axis=1)
x.head()
```

```python
y = data['MEDV']
y.head()
```

```python
from sklearn.preprocessing import StandardScaler

sc_x = StandardScaler()
x = sc_x.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
print (x_train.shape, y_train.shape)
print (x_test.shape, y_test.shape)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rf= RandomForestRegressor(100)
rf= rf.fit(x_train, y_train)
```

```python
predictions = rf.predict(x_test)
```

```python
predictions.shape
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, predictions)
```

## <font color = "green">K-Fold Cross validation</font>

```python
from sklearn.model_selection import KFold
kfold = KFold(n_splits=5, shuffle=False)

for train_index, test_index in kfold.split(x):
    print("Train - \n", train_index, "\nTest - \n", test_index)
    print("=======\n")
# kfold.split(x)
```

```python
rf1= RandomForestRegressor()

n_est = [100, 200, 500]

for n_estimator in n_est:       
    error = []
    for train_index, test_index in kfold.split(x):
        
        x_train, x_test = x[train_index], x[test_index]
        y_train, y_test = y[train_index], y[test_index]
    
        rf1 = RandomForestRegressor(n_estimator)

        rf1 = rf1.fit(x_train, y_train)

        prediction = rf1.predict(x_test)
        
        error.append(mean_squared_error(y_test, prediction))
    print("Results for n_estimator: ", n_estimator)
    print(error)
    print("Average Error = ", sum(error)/len(error))
```
