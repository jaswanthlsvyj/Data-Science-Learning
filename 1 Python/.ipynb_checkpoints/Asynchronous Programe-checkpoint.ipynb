Asynchronous programming is a typer of parallel programming in which a unit of work is allowd to run separately from the primary thread. when the work is complete, it notifies the main thread about completion or failure of the worker thread. There are numerous benefits to using it, such as improved application performance and enhanced responsiveness.

# Multi Processing

```python
from multiprocessing import Process
```

```python
def showSqr(num=2):
    print(num**2)
```

```python
# list of all the processes
procs = []
```

```python
for i in range(5):
    procs.append(Process(target=showSqr, args=(i+1, )))
```

```python
for proc in procs:
    proc.start()
    
print('hello')
```

```python
for proc in procs:
    proc.join()
```

```python
print(procs)
```

# Multi Threading

```python
from threading import Thread
```

```python
def square(n):
    print("square is",n**2)
    
def cube(n):
    print("cube is",n**3)
```

```python
t1 = Thread(target = square, args=(4,))
t2 = Thread(target = cube , args=(3,))
```

- This thread.start() runs the threads prallelly where the output will be asynchronous i.e o/ps will be out of order

```python
t1.start()
t2.start()
print('Hello')
```

- tread.join() will join the o/p synchronouly

```python
from queue import Queue
```

```python
def producer(q):
    for i in range(5):
        q.put(i)
        print('published',i)
        
def consumer(q):
    while True:
        data = q.get()
        print('consumed',data)
        
        
q = Queue()
```

```python
producer_thread = Thread(target = producer, args=(q,))
consumer_thread = Thread(target = consumer, args=(q,))
```

```python
consumer_thread.start()
producer_thread.start()
```

```python
producer_thread.join()
```

```python
consumer_thread.join()
```
